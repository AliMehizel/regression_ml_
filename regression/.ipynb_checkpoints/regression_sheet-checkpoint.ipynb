
## 1. gradient descent 

```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot  as plt
```

```python
from gradient_descent import *
from lineair_regression import LineairRegression

g = SGDRegressor()
data = pd.read_excel('data.xlsx')
Y = data['CONSOMMATION']
X = data[['REVENU','INDICE_PRIX']]

g.fit(X,Y,np.array([12,1,1]),learning_rate=0.00001,max_iter=50)



params = g.params
params
```

```python
one = np.ones(80,dtype=int).reshape(80,1)
y = np.concatenate((one,X.values),1) @ params
pd.Series(y)
```

```python

import matplotlib.pyplot as plt
import numpy as np



# Plot scatter plot

plt.scatter(X['REVENU'].values, Y.values, label='Data')
# Plot lie

plt.plot(X['REVENU'].values, y, color='red', label='Line of Best Fit')
# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot with Line of Best Fit')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
